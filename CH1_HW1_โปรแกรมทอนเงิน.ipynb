```python
while True:
    try:
        A = float(input("กรุณากรอกมูลค่าสินค้า: "))
        break
    except ValueError:
        print("กรุณาใส่จำนวนแบบตัวเลขเท่านั้น!")

while True:
    try:
        B = float(input("กรุณากรอกจำนวนเงินที่รับจากลูกค้า: "))
        if B < A:
            print("จำนวนเงินไม่พอ")
        else:
            break
    except ValueError:
        print("กรุณาใส่จำนวนแบบตัวเลขเท่านั้น!")

if B == A:
    print("รับเงินพอดี ไม่ต้องทอน")
else:
    change = B - A
    print(f"เงินทอน: {change} บาท")


```
